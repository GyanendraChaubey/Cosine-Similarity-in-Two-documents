```python
import pandas as pd
import nltk
```

```python
nltk.download('punkt')
```

```python
data1=pd.read_csv('/content/drive/MyDrive/concat1.csv', encoding= 'unicode_escape')
```

```python
data1.head()
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
stop = open("/content/drive/MyDrive/english.stop.txt","r+")
```

```python
stopword_list=stop.readlines()
```

```python
stopword_list
```

```python
#Removing \n from each stopwords
stopwords=[]
for i in stopword_list:
  stopwords.append(i[:-1])

print(stopwords)
```

```python
sentences=[i for i in data1['VIDEO_TAG'].apply(str)]
#print(sentences)
corpus=[]
for i in range(len(sentences)):
  #Step1: Tokenisation 
  words=nltk.word_tokenize(sentences[i])
  #Step2: Stemming of words 
  words=[stemmer.stem(word) for word in words]
  #Step3: Removing Stop words
  words=[stemmer.stem(word) for word in words if not word in stopwords]
  corpus.append(' '.join(words))
```

```python
print(words)
```

```python
print(corpus)
```

#Bigram

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv1=CountVectorizer(binary=True,ngram_range=(1,2))
```

```python
cv1.fit_transform(corpus).toarray()
```

#Category Discriminating Terms (CDTs)

```python
vocabolaries1=cv1.vocabulary_
vocabolaries1
```

```python
#Setting the threeshold of 100
vocabolaries1 = dict((k, v) for k, v in vocabolaries1.items() if v >= 100)
```

```python
vocabolaries1
```

```python
count1=cv1.fit_transform(corpus).toarray()
```
